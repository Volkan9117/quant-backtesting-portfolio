Ce mini-projet a pour but de maîtriser la manipulation de données financières historiques en utilisant Python. Nous allons nous concentrer sur trois géants de la tech (GAFAM) : <strong>Microsoft</strong>, <strong>Google</strong> et <strong>Amazon</strong>.

<h2 style="color: black;">Objectifs du Défi</h2>
L'objectif principal est d'importer, de structurer et d'inspecter les données boursières sur la période allant de <strong>2022 à 2024</strong>.

Les étapes clés de ce notebook sont :
1.  <strong>Importation :</strong> Récupérer les données pour les tickers `MSFT`, `GOOGL`, et `AMZN`.
2.  <strong>Structuration :</strong> Créer trois DataFrames distincts pour chaque entreprise.
3.  <strong>Exploration :</strong>
    * Afficher les 10 premières lignes (aperçu).
    * Compter les valeurs manquantes (nettoyage).
    * Décrire les statistiques des colonnes.

<hr>
<strong>Outils utilisés :</strong> Python, Pandas, Jupyter Notebook.


```python
# 1. Définition des paramètres
start = "2022-01-01"
end = "2024-12-31"
tickers = ['MSFT', 'GOOGL', 'AMZN']
```

```python
# 2. Importation des librairies
!pip install yfinance
import yfinance as yf
import pandas as pd
```

```python
# 3. Téléchargement des données
data = yf.download(tickers, start=start, end=end)
# 4. Affichage
data.head()
```

```python
#4 Regroupons les tickers:
data_group = yf.download(tickers, start=start, end=end, group_by ="tickers")
data_group.head()
```

```python
# Créer trois DataFrames distincts pour chaque entreprise
MSFT = data_group["MSFT"]
AMZN = data_group["AMZN"]
GOOGL = data_group["GOOGL"]
```

```python
# Afficher les 10 premières lignes "MSFT"
MSFT.head(10)
```

```python
# Afficher les 10 premières lignes "AMZN"
AMZN.head(10)
```

```python
# Compter les valeurs manquantes "GOOGL"
GOOGL.isnull().sum()
```

```python
# Compter les valeurs manquantes "AMZN"
AMZN.isnull().sum()
```

```python
# Compter les valeurs manquantes "MSFT"
MSFT.isnull().sum()
```
